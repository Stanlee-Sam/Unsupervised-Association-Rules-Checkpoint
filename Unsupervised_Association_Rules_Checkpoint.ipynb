```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
#convert to a one-hot encoded DataFrame
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(toy_dataset).transform(toy_dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
#run apriori
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
frequent_itemsets
```

```python
#generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

#  Association Rule Mining - Customer Purchase History Analysis

##  Objective
This project applies **Unsupervised Association Rule Mining (Apriori Algorithm)** to a supermarkets historical purchase dataset.  
The goal is to uncover hidden item relationships that can guide **marketing strategies, product placement, and inventory decisions**.

---

##  Dataset Overview
- Source: Kaggle  *Customer Purchase History Dataset*
- Each row represents one transaction.
- Each column represents a purchased item slot (up to 20 items per transaction).
- Total transactions: 7,500

---

##  Methodology
1. **Data Cleaning**
   - Converted each row into a list of purchased products.
   - Removed missing (`NaN`) values.
   - Prepared a transactional dataset suitable for the Apriori algorithm.

2. **Data Transformation**
   - Used `TransactionEncoder` from `mlxtend` to one-hot encode transactions.

3. **Apriori Algorithm**
   - Minimum Support: `0.02`
   - Minimum Confidence: `0.2`
   - Generated frequent itemsets and association rules.

4. **Evaluation Metrics**
   - **Support:** Frequency of occurrence in transactions.
   - **Confidence:** Probability of buying Y given X.
   - **Lift:** Strength of association (Lift > 1 = positive relation).

---

##  Key Findings

###  Frequent Items
Top commonly purchased items:
- Mineral water  
- Milk  
- Ground beef  
- Frozen vegetables  
- Spaghetti  

###  Strongest Association Rules

| Rule | Confidence | Lift | Interpretation |
|------|-------------|------|----------------|
| Soup  Mineral Water | 0.46 | **1.92** | Soup buyers are almost twice as likely to buy mineral water. |
| Olive Oil  Mineral Water | 0.42 | **1.75** | Health-conscious customers buy both together. |
| Ground Beef  Mineral Water | 0.42 | **1.75** | Meat buyers tend to add mineral water. |
| Spaghetti  Ground Beef | 0.23 | **2.29** | Classic meal combination. |
| Frozen Vegetables  Milk | 0.25 | **1.91** | Breakfast or smoothie pairing. |
| Burgers  Eggs | 0.33 | **1.84** | Typical meal-prep combination. |

---

##  Customer Behavior Insights

1. **Health-Conscious Shoppers**
   - Frequently purchase olive oil, frozen vegetables, chicken, and mineral water.
   - Represent a growing health-focused segment.

2. **Meal-Planning Shoppers**
   - Buy spaghetti + ground beef or frozen vegetables + milk.
   - Likely cooking full meals at home.

3. **Snack & Convenience Shoppers**
   - Purchase cake, chocolate, and mineral water together.
   - Prefer quick and easy food options.

---

##  Business Recommendations

###  Product Placement & Layout
- Place **mineral water near soups, meats, and frozen food sections**.
- Group **spaghetti and ground beef** for convenient meal combo shopping.
- Position **milk near frozen vegetables** to encourage breakfast bundle sales.

###  Promotions
- Create **bundle discounts** (e.g., Buy Spaghetti & Get 10% off Ground Beef).
- Launch **Healthy Meal Packs** (olive oil + frozen vegetables + mineral water).
- Use **checkout recommendations**: suggest mineral water with protein purchases.

###  Inventory Management
- Maintain high stock levels for **mineral water, milk, ground beef**, and other high-lift pairs.
- Reorder complementary products together to support promotions.

###  Customer Segmentation
- **Health Segment:** Olive oil, chicken, frozen veggies, mineral water.  
- **Meal Planners:** Spaghetti, beef, milk.  
- **Snack Buyers:** Cake, chocolate, eggs.

---

##  Expected Impact
- Increased cross-selling and total revenue.  
- Better shelf organization for easier shopping.  
- Smarter inventory control through data-driven insights.

---

##  Conclusion
Association Rule Mining successfully revealed valuable shopping patterns from supermarket transactions.  
By leveraging these insights, supermarket owners can design **targeted promotions, improve store layout**, and **optimize inventory**, driving both customer satisfaction and profitability.
